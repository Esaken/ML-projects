```python

req = requests.get("https://www.centralbank.go.ke/")
soup = BeautifulSoup(req.text, 'html.parser')

links = soup.findAll('tr', class_="tg-4eph")


formatted_links = []

for link in links:
    data = {
        'tg-4eph': link.find_all('td')[0].a.text, 
        'tg-4eph': link.find_all('td')[1].a.text
        
    } 
    formatted_links.append(data)

print(formatted_links)
```
